```python
import pandas as pd

file_path = 'population_csv.csv'  

data = pd.read_csv(file_path)

print(data.head())  
print(data.info())  
```

```python
import matplotlib.pyplot as plt

# Example: Plotting a bar chart for a subset of data
subset_data = data.sample(10)  # Taking a random sample of 10 rows for visualization

plt.figure(figsize=(10, 6))
plt.bar(subset_data['Country Name'], subset_data['Value'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Sample Bar Chart: Country Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'Value' column represents population
plt.figure(figsize=(10, 6))
plt.hist(data['Value'], bins=30, color='skyblue')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Population Distribution Histogram')
plt.grid(True)
plt.tight_layout()
plt.show()
```
